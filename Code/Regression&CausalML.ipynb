```python
#Creating a dataframe with both excel files
import pandas as pd
import datetime as dt
# File paths
csv_file_path = "Vader_Posts_sentiment_scores_Final.csv"
vector_file_path = "one_three_grams_count_tfidf_lemmatized.csv"
excel3_file_path = "blackfriday_topics CSV.csv"

# Step 1: Load the Excel Files
df1 = pd.read_csv(csv_file_path)
df2 = pd.read_csv(vector_file_path)
df3 = pd.read_csv(excel3_file_path, sep=",")
df2.drop(['score','subreddit_id','var_year'], axis=1, inplace=True)
df3.drop(['Unnamed: 0', 'retrieved_on', 'body', 'Topic', 'Topic_0_Prob', 'Topic_1_Prob', 'Shopping_Type'], axis=1, inplace=True)
```

```python
merged_df = pd.concat([df1, df2, df3], axis=1)
merged_df.columns
```

```python
### Modify date 
merged_df['retrieved_on'] = pd.to_datetime(merged_df['retrieved_on'])
merged_df['Year_var'] = merged_df['retrieved_on'].dt.year
merged_df.head()

merged_df['Year_var']= merged_df['Year_var'] - 2017

df = merged_df
df.drop(['body'], axis=1, inplace=True)
df
```

```python
###Encoding subreddit_id

# Assuming you have a DataFrame named df with a column called 'subreddit_id'
# Let's create dummy variables for the 'subreddit_id' column
dummies = pd.get_dummies(df['subreddit_id'], prefix='subreddit', dtype=float)

# Concatenate the original DataFrame with the dummy variables
df = pd.concat([df, dummies], axis=1)

# Drop the original 'subreddit_id' column if needed
df.drop(['subreddit_id', 'retrieved_on'], axis=1, inplace=True)

# Display the DataFrame with dummy variables
df['Sentiment'] = df['Sentiment'].map({'Positive': 1, 'Negative': 0, 'Neutral': 2})
# drop rows where sentiment is neutral
df = df[df['Sentiment'] != 2]
df.head()
```

```python
# check class imbalance
df['Sentiment'].value_counts()

# Upsample minority class
from sklearn.utils import resample

# Separate majority and minority classes
df_majority = df[df.Sentiment==1]
df_minority = df[df.Sentiment==0]

# Upsample minority class
df_minority_upsampled = resample(df_minority, 
                                 replace=True,     # sample with replacement
                                 n_samples=10000,    # to match majority class
                                 random_state=123) # reproducible results

df_upsampled = pd.concat([df_majority, df_minority_upsampled])
df_upsampled
```

```python
df_reg = df_upsampled
df_upsampled = df_upsampled.drop('Sentiment Score', axis=1)
```

```python
from sklearn.linear_model import LogisticRegression

X = df_upsampled.drop('Sentiment', axis=1)
y = df_upsampled['Sentiment']
```

```python
from sklearn.model_selection import train_test_split

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)  # 80% training, 20% testing

features_to_scale_1 = X_train[['Year_var', 'score']]
features_to_scale_2 = X_test[['Year_var', 'score']]

from sklearn.preprocessing import MinMaxScaler

scaled_features_1 = MinMaxScaler().fit_transform(features_to_scale_1)
scaled_features_2 = MinMaxScaler().fit_transform(features_to_scale_2)

X_train[['Year_var', 'score']] = scaled_features_1
X_test[['Year_var', 'score']] = scaled_features_2

# Create and fit the linear regression model on the training set
model = LogisticRegression(n_jobs=-1, max_iter=1000, random_state=0)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print('Accuracy:', accuracy_score(y_test, y_pred))
print('Precision:', precision_score(y_test, y_pred))
print('Recall:', recall_score(y_test, y_pred))
print('F1 Score:', f1_score(y_test, y_pred))
```
